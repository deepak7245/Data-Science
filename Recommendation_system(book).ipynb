```python
Book = Book.groupby(['User_ID', 'Book_Title'])['Book_Rating'].mean().reset_index()
```

```python
User = Book.pivot(index= 'User_ID', columns= 'Book_Title', values= 'Book_Rating')
```

```python
#impute those NANs with '0' values

User.fillna(0, inplace=True)
```

```python
#finding similer kind of Users by metrics of algorithm is pairwise_distances
# and calculating Cosine 

from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
User_1 = 1 - pairwise_distances(User.values,metric= 'cosine')
```

```python
User_1
```

```python
#Store the results in a dataFrame

user_users = pd.DataFrame(User_1)
```
